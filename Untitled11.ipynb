```python
import pandas as pd

# path
file_path = '/content/drive/My Drive/KNPS/bass/2024_Bass_Survey_Songnisan.csv'

# CSV read (encoding)
data = pd.read_csv(file_path, encoding='cp949')
filtered_data = data[~data['Survey_Round'].str.contains('R001|R002')]
print(filtered_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for publication-quality visuals
sns.set_theme(style="whitegrid", context="paper", font_scale=1.5)

# Custom color palette for Nature-style visuals
nature_colors = ["#4E79A7", "#F28E2B", "#E15759"]

# Plotting Length Distribution
plt.figure(figsize=(8, 6))
sns.histplot(filtered_data['Length'], bins=20, kde=True, color=nature_colors[0], edgecolor='black')
plt.title('Length Distribution', fontsize=16)
plt.xlabel('Length (cm)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()
plt.show()

# Plotting Weight Distribution
plt.figure(figsize=(8, 6))
sns.histplot(filtered_data['Weight'], bins=20, kde=True, color=nature_colors[1], edgecolor='black')
plt.title('Weight Distribution', fontsize=16)
plt.xlabel('Weight (g)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()
plt.show()

# Plotting Surface Temperature Distribution
plt.figure(figsize=(8, 6))
sns.histplot(filtered_data['Surface_Temperature'], bins=20, kde=True, color=nature_colors[2], edgecolor='black')
plt.title('Surface Temperature Distribution', fontsize=16)
plt.xlabel('Surface Temperature (Â°C)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Calculating correlation matrix
correlation_matrix = filtered_data[['Length', 'Weight', 'Surface_Temperature']].corr()

# Setting up a style for publication-quality heatmap
plt.figure(figsize=(8, 6))
sns.set_theme(style="white", context="paper", font_scale=1.5)

# Drawing the heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,                # Show correlation values
    fmt=".2f",                 # Format for annotations
    cmap="vlag",               # Sophisticated colormap
    cbar_kws={'shrink': 0.8},  # Adjust colorbar size
    linewidths=0.5,            # Add subtle grid lines
    square=True                # Ensure square cells
)

# Adding title and axis labels
plt.title('Correlation Matrix: Length, Weight, Surface Temperature', fontsize=16, pad=20)
plt.xticks(fontsize=12, rotation=45, ha="right")
plt.yticks(fontsize=12, rotation=0)
plt.tight_layout()
plt.show()
```

```python
# ì²´íì§ì ê³ì°
filtered_data['Condition_Factor'] = (filtered_data['Weight'] / filtered_data['Length']**3) * 100000

# ì²´íì§ìì íì¸µìì¨ì ì°ì ë ê·¸ë¦¬ê¸°
plt.figure(figsize=(8, 6))
sns.regplot(
    x='Surface_Temperature',
    y='Condition_Factor',
    data=filtered_data,
    scatter_kws={'alpha': 0.7},
    line_kws={'color': 'red'}
)
plt.title('Condition Factor vs. Surface Temperature', fontsize=16)
plt.xlabel('Surface Temperature (Â°C)', fontsize=14)
plt.ylabel('Condition Factor', fontsize=14)
plt.tight_layout()
plt.show()

# ìê´ê³ì ê³ì°
correlation = filtered_data[['Condition_Factor', 'Surface_Temperature']].corr().iloc[0,1]
print(f"Correlation between Condition Factor and Surface Temperature: {correlation:.2f}")
```
